```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Carregar o dataset
df = pd.read_excel('DBQ4_TCC2.xlsx', sheet_name='DB', date_parser='Data Hora')  # Substitua 'seu_arquivo.csv' pelo nome do seu arquivo

```

```python
df.head()
```

```python
# Verificar as primeiras linhas do dataset
print(df.head())

# Verificar informações gerais do dataset
print(df.info())

# Converter colunas de Data e Hora para o tipo adequado
df['Data Hora'] = pd.to_datetime(df['Data Hora'])

# Análise temporal das vendas
df['Data'] = df['Data Hora'].dt.date
vendas_por_data = df.groupby('Data').agg({'CupomFiscal': 'nunique', 'Valor': 'sum'})

```

```python
# Plotar gráfico de vendas ao longo do tempo
plt.figure(figsize=(12, 6))
vendas_por_data['Valor'].plot(kind='line', marker='o')
plt.title('Vendas ao longo do tempo')
plt.xlabel('Data')
plt.ylabel('Valor das Vendas')
plt.grid(True)
plt.show()

```

```python
# Análise dos produtos mais vendidos
produtos_mais_vendidos = df['Produto'].value_counts().head(10)

# Plotar gráfico dos produtos mais vendidos
plt.figure(figsize=(10, 6))
produtos_mais_vendidos.plot(kind='bar')
plt.title('Top 10 Produtos Mais Vendidos')
plt.xlabel('Produto')
plt.ylabel('Quantidade Vendida')
plt.xticks(rotation=45)
plt.show()

```

```python

# Análise de correlação entre quantidade e valor
correlacao_quantidade_valor = df[['Quantidade', 'Valor']].corr()

# Plotar heatmap da correlação
plt.figure(figsize=(6, 4))
sns.heatmap(correlacao_quantidade_valor, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlação entre Quantidade e Valor')
plt.show()
```

```python
# Converter colunas de Data e Hora para o tipo adequado
df['Data Hora'] = pd.to_datetime(df['Data Hora'])

# Extrair o dia da semana e o mês
df['Dia_da_Semana'] = df['Data Hora'].dt.day_name()
df['Mes'] = df['Data Hora'].dt.month

# Calcular o valor total das vendas por produto, dia da semana e mês
vendas_por_produto_dia_da_semana = df.groupby(['Produto', 'Dia_da_Semana'])['Valor'].sum()
vendas_por_produto_mes = df.groupby(['Produto', 'Mes'])['Valor'].sum()

# Escolha de um produto específico para análise
produto_escolhido = '56 - BRAZUCA'  # Substitua pelo nome do produto que deseja analisar

# Plotar gráfico de barras do valor das vendas desse produto por dia da semana
plt.figure(figsize=(12, 6))
vendas_por_produto_dia_da_semana[produto_escolhido].plot(kind='bar', stacked=True)
plt.title(f'Vendas do Produto "{produto_escolhido}" por Dia da Semana')
plt.xlabel('Dia da Semana')
plt.ylabel('Valor das Vendas')
plt.xticks(rotation=45)
plt.legend(title='Produto')
plt.show()
```

```python
# Calcular o valor total das vendas por produto e dia da semana
vendas_por_produto_dia_da_semana = df.groupby(['Produto', 'Dia_da_Semana'])['Valor'].sum()

# Encontrar os 10 produtos mais vendidos
produtos_mais_vendidos = df['Produto'].value_counts().head(10).index.tolist()

# Encontrar o dia de maior venda para os 10 produtos mais vendidos
for produto in produtos_mais_vendidos:
    vendas_produto = vendas_por_produto_dia_da_semana.loc[produto]
    dia_maior_venda = vendas_produto.idxmax()  # Encontrar o dia de maior venda
    valor_maior_venda = vendas_produto.max()  # Valor da maior venda
    print(f"Produto: {produto} - Dia de Maior Venda: {dia_maior_venda} - Valor: {valor_maior_venda}")
```

```python
# Calcular o valor total das vendas por produto e dia da semana
vendas_por_produto_dia_da_semana = df.groupby(['Produto', 'Dia_da_Semana'])['Valor'].sum()

# Encontrar os 10 produtos mais vendidos
produtos_mais_vendidos = df['Produto'].value_counts().head(10).index.tolist()

# Filtrar o DataFrame apenas para os 10 produtos mais vendidos
df_top_10 = df[df['Produto'].isin(produtos_mais_vendidos)]

# Calcular o total de vendas por dia da semana para os 10 produtos mais vendidos
vendas_por_dia_produto = df_top_10.groupby(['Dia_da_Semana', 'Produto'])['Valor'].sum().unstack()

# Plotar gráfico de barras empilhadas
plt.figure(figsize=(20,10))
vendas_por_dia_produto.plot(kind='bar', stacked=True)
plt.title('Contribuição dos 10 Produtos Mais Vendidos por Dia da Semana')
plt.xlabel('Dia da Semana')
plt.ylabel('Valor das Vendas')
plt.xticks(rotation=45)
plt.legend(title='Produto', prop={'size': 6})
plt.show()
```

```python
# Converter colunas de Data e Hora para o tipo adequado
df['Data Hora'] = pd.to_datetime(df['Data Hora'])

# Extrair o mês e o ano
df['Ano'] = df['Data Hora'].dt.year
df['Mes'] = df['Data Hora'].dt.month

# Calcular o valor total das vendas por produto, ano e mês
vendas_por_produto_ano_mes = df.groupby(['Produto', 'Ano', 'Mes'])['Valor'].sum()

# Encontrar os 10 produtos mais vendidos
produtos_mais_vendidos = df['Produto'].value_counts().head(10).index.tolist()

# Filtrar o DataFrame apenas para os 10 produtos mais vendidos
df_top_10 = df[df['Produto'].isin(produtos_mais_vendidos)]

# Calcular o total de vendas por ano, mês e produto para os 10 produtos mais vendidos
vendas_top_10_produtos = df_top_10.groupby(['Produto', 'Ano', 'Mes'])['Valor'].sum().unstack(level=0)

# Plotar gráfico de vendas ao longo do tempo para os 10 produtos mais vendidos
plt.figure(figsize=(12, 6))
for produto in vendas_top_10_produtos.columns:
    vendas_top_10_produtos[produto].plot(label=produto, linewidth=2)

plt.title('Vendas ao Longo do Tempo para os 10 Produtos Mais Vendidos')
plt.xlabel('Ano/Mês')
plt.ylabel('Valor das Vendas')
plt.legend(title='Produto')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Calcular a quantidade de produtos por Cupom Fiscal
quantidade_por_cupom = df.groupby('CupomFiscal')['Produto'].count().reset_index()
quantidade_por_cupom.columns = ['CupomFiscal', 'Quantidade_Produtos']

# Criar um gráfico de boxplot para a distribuição da quantidade de produtos por Cupom Fiscal
plt.figure(figsize=(8, 3))
sns.boxplot(x='Quantidade_Produtos', data=quantidade_por_cupom, orient="Vertical")
plt.title('Distribuição da Quantidade de Produtos por Cupom Fiscal')
plt.ylabel('Quantidade de Produtos por Cupom')
plt.grid(axis='y')  # Adicionar linhas de grade no eixo y para visualização da distribuição
plt.show()
```

```python
# Calcular a soma do valor total por Cupom Fiscal
valor_por_cupom = df.groupby('CupomFiscal')['Valor'].sum().reset_index()
valor_por_cupom.columns = ['CupomFiscal', 'Valor_Total']

# Calcular a quantidade de produtos por Cupom Fiscal
quantidade_por_cupom = df.groupby('CupomFiscal')['Produto'].count().reset_index()
quantidade_por_cupom.columns = ['CupomFiscal', 'Quantidade_Produtos']

# Juntar os dataframes para ter a quantidade de produtos e o valor total por cupom
cupom_info = pd.merge(quantidade_por_cupom, valor_por_cupom, on='CupomFiscal')

# Criar um gráfico de boxplot com a soma do valor total por Cupom Fiscal destacada
plt.figure(figsize=(8, 3))
sns.boxplot(x='Valor_Total', data=cupom_info, boxprops=dict(alpha=0.5))  # Boxplot da quantidade de produtos
#sns.stripplot(y='Valor_Total', data=cupom_info, color='red', alpha=0.5)  # Pontos destacando a soma do valor total
plt.title('Distribuição de Valores de Cupom')
#plt.ylabel('Quantidade de Produtos por Cupom')
plt.xlabel('Soma do Valor Total por Cupom')
plt.grid(axis='y')  # Adicionar linhas de grade no eixo y para visualização da distribuição
plt.show()
```
